```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# total enrollment
te = pd.read_csv("data/totalenrollment.csv", index_col='UnitID')
```

```python
te.head()
```

```python
# reverse order of columns up to the first column
te_reversed = te.iloc[:,:1:-1]
te_reversed.insert(0,'Institution',te['Institution'])
te = te_reversed
```

```python
def get_enrollment(institution_name):
    institution_data = te[te['Institution'] == institution_name]
    return institution_data

# Test the function with an example
get_enrollment('University of Phoenix-Arizona')
```

```python
def top10_by_year(year):
    top10_institutions = te[['Institution', str(year)]].sort_values(by=str(year), ascending=False).head(10)
    return top10_institutions

# Test the function with an example
top10_by_year(2011)
```

```python
top10_by_year(2011)
```

```python
get_enrollment('University of the Rockies')
```

```python
def plot_enrollment(institution_name):
    institution_data = get_enrollment(institution_name)
    institution_data.drop('Institution', axis=1).T.plot(kind='line', title=institution_name, legend=False)
    plt.xlabel('Year')
    plt.ylabel('Enrollment')
    plt.grid()
    plt.show()

# Test the function with an example
plot_enrollment('University of Phoenix-Arizona')
```

```python
def plot_enrollment(*institution_names):
    plt.figure(figsize=(10, 6))
    for institution_name in institution_names:
        institution_data = get_enrollment(institution_name)
        plt.plot(institution_data.drop('Institution', axis=1).columns, institution_data.drop('Institution', axis=1).values[0])
    plt.xlabel('Year')
    plt.ylabel('Enrollment')
    plt.title('Enrollment Over Time')
    plt.legend(institution_names)
    plt.show()

# Test the function with an example
plot_enrollment('University of Phoenix-Arizona', 'Ashford University','Western Governors University','Southern New Hampshire University')
```

```python
def plot_enrollment(*institution_names):
    plt.figure(figsize=(10, 6))
    for institution_name in institution_names:
        institution_data = get_enrollment(institution_name)
        sns.lineplot(x=institution_data.drop('Institution', axis=1).columns, y=institution_data.drop('Institution', axis=1).values[0])
    plt.xlabel('Year')
    plt.ylabel('Enrollment')
    plt.title('Enrollment Over Time')
    plt.legend(institution_names)
    plt.grid()
    plt.show()

# Test the function with an example
# Test the function with an example
plot_enrollment('University of Phoenix-Arizona', 'Ashford University','Western Governors University','Southern New Hampshire University')
```
