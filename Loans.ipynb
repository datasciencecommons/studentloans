```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencecommons/studentloans/main/data/fedloans.csv'
loans = pd.read_csv(url, index_col='UnitID')
loans.head()
```

```python
loans_reversed = loans.iloc[:,:1:-1]
loans_reversed.insert(0,'Institution',loans['Institution'])
loans = loans_reversed
```

```python
loans.head()
```

```python
loans['Total'] = loans.iloc[:, 1:].sum(axis=1)
loans.head()
```

```python
loan_totals = loans[['Institution', 'Total']]
loan_totals = loan_totals.sort_values('Total', ascending=False)
loan_totals.head(10)
```

```python
loan_totals['Mill'] = loan_totals['Total'] / 1e6
loan_totals['Bill'] = loan_totals['Total'] / 1e9
loan_totals.head(25)
```

```python
import matplotlib.pyplot as plt

def plot_institution(institution):
    institution_data = loans[loans['Institution'] == institution]
    institution_data = institution_data.drop(columns=['Institution', 'Total']).T
    institution_data.columns = ['Bill']
    institution_data['Bill'] = institution_data['Bill'] / 1e9
    institution_data.plot(kind='line', title=institution)
    plt.xlabel('Year')
    plt.ylabel('Billions of Dollars')
    plt.show()
```

```python
plot_institution('Grand Canyon University')
```

```python
def trend_institution(institution):
    institution_data = loans[loans['Institution'] == institution]
    institution_data = institution_data.drop(columns=['Institution', 'Total']).T
    institution_data.columns = ['Loan Amount']
    institution_data['Loan Amount'] = institution_data['Loan Amount'] / 1e6
    return institution_data
```

```python
trend_institution('Grand Canyon University')
```
